# using combaine model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

tree=DecisionTreeClassifier(random_state=20)
svm=SVC(kernel="sigmoid",gamma=0.1,C=1)#kernel='linear', 'poly', 'rbf', 'sigmoid', 'precomputed' 
                         #  any one of it
knn=KNeighborsClassifier(n_neighbors=5,metric='euclidean')
logreg=LogisticRegression(random_state=10)
```

```python
# now creating a list with all this model
models=[tree,svm,knn,logreg]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

for model in models:
    model.fit(X_train,Y_train) # fit model
    Y_pred=model.predict(X_test)# then predict on the test set
    accuracy=accuracy_score(Y_test,Y_pred)
    clf_report= classification_report(Y_test,Y_pred)
    print(confusion_matrix(Y_test,Y_pred))
    print("The accuracy of the", type(model).__name__," model is" , accuracy*100)
    print("Classifeication report: \n", clf_report)
    print("\n")
```

# Used to tell the mobel is best or not and by change some randomstate and estimators we can find whis is the best mobel 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier

# create the sub models
estimators = []
#model1 = LogisticRegression()
#estimators.append(('log', model1))
model2 = DecisionTreeClassifier(criterion='gini',random_state=10)
estimators.append(('cart', model2))
model3 = SVC(kernel="rbf", C=90,gamma=0.1)
estimators.append(('svm', model3))
#model4 = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
#estimators.append(('knn', model4))


# create the ensemble model
ensemble = VotingClassifier(estimators)
ensemble.fit(X_train,Y_train)
Y_pred=ensemble.predict(X_test)
#print(Y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(Y_test,Y_pred))
print(accuracy_score(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```

# the combation of mobel and tell which is the best mobel accurecy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import StackingClassifier
from sklearn.neighbors import KNeighborsClassifier

# create the sub models
level0 = []
model1 = LogisticRegression()
level0.append(('log', model1))
model2 = DecisionTreeClassifier(criterion='gini',random_state=10)
level0.append(('cart', model2))
model3 = SVC(kernel="rbf", C=90,gamma=0.1)
level0.append(('svm', model3))
model4 = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
level0.append(('knn', model4))

level1=LogisticRegression()

# create the ensemble model
ensemble = StackingClassifier(estimators=level0, final_estimator=level1, cv=5,passthrough=True)
ensemble.fit(X_train,Y_train)
Y_pred=ensemble.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(Y_test,Y_pred))
print(accuracy_score(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```
